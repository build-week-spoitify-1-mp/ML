```
# Install Spotify api wrapper spotipy
!pip install spotipy --upgrade
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

def spotify_api(url_list, prebuilt_df=None):

  spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials('bafe55840cf1423384837515f8774ce4', '5f9f58ac3a974f4fa1eb687c453678b5'))
  
  # Call artists, tracks, track_ids, albums, and audio features from spotify
  artists = []
  tracks = []
  tracks_id = []
  albums = []
  audio_features = []
  # Create audio feature lists
  danceability = []
  energy = []
  loudness = []
  speechiness = []
  accousticness = []
  instrumentalness = []
  liveness = []
  valence = []
  tempo = []

  for url in url_list:
    results = spotify.artist_top_tracks(url)
    try:
      for i in range(10):
        tracks.append(results['tracks'][i]['name'])
        tracks_id.append(results['tracks'][i]['id'])
    except:
      
      continue

  for track in tracks_id:
    artists.append(spotify.track(track)['artists'][0]['name'])
    albums.append(spotify.track(track)['album']['name'])
    audio_features.append(spotify.audio_features(track))

  for track in audio_features:
    danceability.append(track[0]['danceability'])
    energy.append(track[0]['energy'])
    loudness.append(track[0]['loudness'])
    speechiness.append(track[0]['speechiness'])
    accousticness.append(track[0]['acousticness'])
    instrumentalness.append(track[0]['instrumentalness'])
    liveness.append(track[0]['liveness'])
    valence.append(track[0]['valence'])
    tempo.append(track[0]['tempo'])

  # # Input our data into a dataframe
  df = pd.DataFrame({'Artist':artists, 'Track_name':tracks, 'track_id':tracks_id,
                            'Album':albums, 'Danceability':danceability, 
                            'Energy':energy, 'Loudness':loudness, 
                            'Speechiness':speechiness, 'Accousticness':accousticness, 
                            'Instrumentalness':instrumentalness,
                            'Liveness':liveness, 'Valence':valence, 'Tempo':tempo}
                                    )
  
  df = pd.concat([df, prebuilt_df])

  return df
```
